```python
rule1 = data['Species'].isin(['setosa', 'versicolor', 'virginica'])
```

```python
rule2 = (data[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']] > 0).all(axis=1)
```

```python
Rules = pd.DataFrame({'Rule 1': rule1, 'Rule 2': rule2})
```

```python
rule3 = data['Petal.Length'] >= 2 * data['Petal.Width']
```

```python
rule4 = data['Sepal.Length'] <= 30
```

```python
rule5 = (data['Sepal.Length'] > data['Petal.Length']) & (data['Sepal.Width'] > data['Petal.Width'])
```

```python
Rules = pd.DataFrame({
    "Rule 1": rule1,
    "Rule 2": rule2,
    "Rule 3": rule3,
    "Rule 4": rule4,
    "Rule 5": rule5
})
```

```python
Rules = Rules.astype(int)
print(Rules.head())
```

```python
violations = Rules.apply(lambda x: len(x) - x.sum())
print("Violations per rule:\n", violations)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot violations
plt.figure()
violations.plot(kind='bar')
plt.title("Number of records violating each rule")
plt.xlabel("Rules")
plt.ylabel("Count of Violations")
plt.show()
```
